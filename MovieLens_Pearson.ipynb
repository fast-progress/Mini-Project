```python
import pandas as pd
```

```python
titles =pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
```

```python
titles.head(10)
```

```python
ratings.head(10)
```

```python
ratings = pd.merge(titles, ratings, on= "movie_id")
```

```python
ratings.head(10)
```

```python
reviews = ratings.groupby('title')['rating'].agg(['count','mean']).reset_index().round(1)
```

```python
reviews.head()
```

```python
movies = pd.crosstab(ratings['user_id'],ratings['title'], values= ratings['rating'], aggfunc = 'sum')
```

```python
titles
```

```python
userInput = ["No Escape (1994)", "Next Karate Kid, The (1994)", "Shadow, The (1994)"]
```

```python
movies.corrwith(movies[userInput[0]], method = 'pearson')
```

```python
similarity = movies.corrwith(movies[userInput[0]], method='pearson') + movies.corrwith(movies[userInput[1]], method='pearson') + movies.corrwith(movies[userInput[2]], method='pearson')
```

```python
correlatedMovies = pd.DataFrame(similarity, columns = ['correlation'])
correlatedMovies = pd.merge(correlatedMovies, reviews, on = 'title')
correlatedMovies = pd.merge(correlatedMovies, titles, on = 'title')
```

```python
correlatedMovies
```

```python
output = correlatedMovies[(correlatedMovies['mean']>3.5) & (correlatedMovies['count']>=100)].sort_values('correlation', ascending = False)
```

```python
output.head(15)
```

```python
del output['movie_id']
del output['correlation']
output.rename(columns={"title":"Movie Suggestions","count": "No. of ratings", "genre":"Genres", "mean":"Ratings"}).head(25)
```
